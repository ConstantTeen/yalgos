```python
IN = -1 
OUT = 1

with open('input.txt') as f:
    n = int(f.readline())
    events = []
    
    for _ in range(n):
        l, r = map(int, f.readline().split())
        events += [(l, IN), (r, OUT)]
    
    events.sort()

cnt = 0
s = 0
last_x = events[0][0]

for x, ev in events:
    if ev == IN:
        if cnt == 0:
            last_x = x
        cnt+= 1
    elif ev == OUT:
        s += x - last_x
        last_x = x
        cnt -= 1

print(s)
```

```python
IN = -1
OUT = 1
CAT = 0

with open('input.txt') as f:
    n, m = map(int, f.readline().split())
    
    events = []
    
    arr = list(map(int, f.readline().split()))
    
    for x in arr:
        events += [(x, CAT, -1)]
        
    for i in range(m):
        l,r = map(int, f.readline().split())
        
        events += [(l, IN, i), (r, OUT, i)]
        
    events.sort()
    
ans = [0 for _ in range(m)]
cnts = [0 for _ in range(m)]
cnt = 0

for x, ev, i in events:
    if ev == IN:
        cnts[i] = -cnt
    elif ev == CAT:
        cnt += 1
    elif ev == OUT:
        ans[i] = cnt + cnts[i]

print(*ans)
```

```python
import math

IN = -1
OUT = 1

with open('input.txt') as f:
    n = int(f.readline())
    r1_max = -1
    r2_min = 1e5
    events = []
    
    for i in range(n):
        r1,r2, phi1, phi2 = map(float, f.readline().split())
        
        if r1 > r1_max:
            r1_max = r1
        if r2 < r2_min:
            r2_min = r2
            
        events += [(phi1, IN, i), (phi2, OUT, i)]
    
if r1_max >= r2_min:
    print(0)
else:
    thickness = (r2_min**2 - r1_max**2)/2
    
    events.sort()
    s = set()
    
    for phi, ev, i in events:
        if ev == IN:
            s.add(i)
        elif ev == OUT:
            s.discard(i)
    
    events += [(2*math.pi, OUT, -1)]
    
    cnt = len(s)
    phi_sum = 0
    last_phi = 0
    
    for phi, ev, i in events:
        if ev == IN:
            last_phi = phi
            cnt += 1
        elif ev == OUT:
            if cnt == n:
                phi_sum += phi - last_phi
            cnt-=1
    
    print(phi_sum*thickness)
```

```python
with open('input.txt') as f:
    m = int(f.readline())
    events = []
    data = []
    i = 0
    
    for row in f:
        row = row.rstrip()
        if len(row) == 0 or row == '0 0':
            break
        
        l, r = map(int, row.split())
        
        if r > l and l <= m and r >= 0:
            data += [(l,r)]
            l,r = max(0, l), min(m, r)
            events += [(l, r-l, i)]
            i+=1
    n = i
    events.sort()

start = 0
max_ = 0
i_ = 0
ans = set()
flag = True

if events[0][0] > 0 or events[-1][0] + events[-1][1] < m:
    flag = False
else:
    for left, length, i in events:
        if left <= start and left + length >= max_:
            i_ = i
            max_ = left + length
        elif left > start and left > max_:
            flag = False
            break
        elif left > start:
            ans.add(data[i_])
            i_ = i
            start = max_
            max_ = left + length
        
        if left + length == m:
                ans.add(data[i])
                break

if flag:
    print(len(ans))
    
    for l,r in sorted(ans):
        print(l,r)
else:
    print('No solution')
```
