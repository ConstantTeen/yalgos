```python
EVENT_OUT = 1
EVENT_IN = -1

with open('input.txt', 'r') as f:
    n,m = map(int, f.readline().split())
    events = []
    
    for _ in range(m):
        b, e = map(int, f.readline().split())
        events += [(b, EVENT_IN), (e, EVENT_OUT)]

events.sort()

cnt = 0
last_p = 0
cnt_in = 0
cnt_out = 0

for p, ev in events:
    if ev == EVENT_IN:
        if cnt_in == cnt_out:
            cnt_in += 1
            cnt+= p - last_p
        elif cnt_in > cnt_out:
            cnt_in += 1
    elif ev == EVENT_OUT: 
        cnt_out+= 1
        
        if cnt_in == cnt_out:
            last_p = p+1
print(cnt + n - last_p)   
```

```python
EVENT_OUT = 1
EVENT_IN = -1
EVENT_NONE = 0

with open('input.txt', 'r') as f:
    n,m = map(int, f.readline().split())
    events = []
    
    for _ in range(n):
        a,b = sorted(map(int, f.readline().split()))
        
        events += [(a, EVENT_IN), (b, EVENT_OUT)]
    
    arrm = list(map(int, f.readline().split()))
    
for e in arrm:
    events += [(e, EVENT_NONE)]
events.sort()

cnts = dict()
cnt = 0

for point, e in events:
    if e == EVENT_NONE:
        cnts[point] = cnt
    elif e == EVENT_IN:
        cnt+=1
    elif e == EVENT_OUT:
        cnt -= 1
res = []

for p in arrm:
    res += [cnts[p]]
    
print(*res)
```

```python
EVENT_OUT = 1
EVENT_IN = -1

with open('input.txt', 'r') as f: 
    n = int(f.readline())
    events = []
    
    for i in range(n):
        in_, out_ = map(int, f.readline().split())
        
        if out_ - in_ >= 5:
            events += [(in_, EVENT_IN, i), (out_-5, EVENT_OUT, i)]
events.sort()        

n = len(events) 

if n == 0:
    print(0, 1, 6)
elif n == 2:
    print(1, events[0][0], events[0][0] + 5)
else:
    s = set()
    max_cnt = 0
    best1, best2 = 0,0
    
    for i in range(n):
        first = events[i]
        if first[1] == EVENT_IN:
            s.add(first[2])
            if len(s) > max_cnt:
                max_cnt = len(s)
                best1, best2 = first[0], first[0] + 5
        cnt2 = 0    
        for j in range(i+1, n):
            second = events[j]
            
            if second[1] == EVENT_IN:
                cnt2 += 1
            if second[0] - first[0] >= 5 and len(s)+cnt2 > max_cnt:
                max_cnt = len(s)+cnt2
                best1, best2 = first[0],second[0]
            if second[1] == EVENT_OUT and second[2] not in s:
                cnt2 -=1   
        if first[1] == EVENT_OUT:
            s.remove(first[2])
        
    print(max_cnt, best1, best2)      
        
            
```

```python
IN = -1
OUT = 1

with open('input.txt', 'r') as f:
    n = int(f.readline())
    events = []
    
    for i in range(n):
        h_start, m_start, h_end, m_end = map(int, f.readline().split())
        
        if (h_start, m_start) > (h_end, m_end):
            events += [((24, 0), OUT), ((0,0), IN)]
            events += [((h_start, m_start), IN), ((h_end, m_end), OUT)]
        elif (h_start, m_start) == (h_end, m_end):
            events += [((24, 0), OUT), ((0,0), IN)]
        else:
            events += [((h_start, m_start), IN), ((h_end, m_end), OUT)]
    
events.sort()

cnt = 0
ans = 0

for time, ev in events:
    if ev == IN:
        cnt+=1
        last_in = time
    elif ev == OUT:
        if cnt == n:
            ans += 60*(time[0] - last_in[0]) + time[1] - last_in[1]
        cnt -= 1
print(ans)
```

```python
import datetime

IN = 1
OUT = -1

with open('input.txt', 'r') as f:
    n = int(f.readline())
    events = []
    
    for i in range(n):
        d_birth, m_birth, y_birth, d_death, m_death, y_death = map(int, f.readline().split())
        
        birth = (y_birth+18,m_birth,d_birth)
        death = min((y_birth+80,m_birth,d_birth), (y_death,m_death,d_death))
        
        if death > birth:
            events += [(birth, IN, i+1), (death, OUT, i+1)]
            
events.sort()

if len(events) == 0:
    print(0)
else:
    s = set()
    for time, ev, i in events:
        if ev == IN:
            s.add(i)
            rest = False
        elif ev == OUT:
            if not rest:
                print(*s)
                rest = True
            s.discard(i)
```

```python
with open('input.txt', 'r') as f:
    m,n = map(int, f.readline().split())
    events = set()
    
    for i in range(n):
        t,z,y = map(int, f.readline().split())
        m_ = m
        last_t = 1
        
        while m_ > 0:
            if (last_t-1) % (z*t+y) == z*t:
                last_t += y
            else:
                m_ -= 1
                last_t += t 
                events.add((last_t, i))

if m == 0:
    print(0)
    print(*[0 for _ in range(n)])
else:
    events = sorted(events)

    cnts = [0 for _ in range(n)]
    m_ = m

    for time, i in events:
        m_ -= 1
        cnts[i] += 1

        if m_ == 0:
            break

    print(time-1)
    print(*cnts)
```
